```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def get_name_by_num(num):
    for doc in documents:
        if doc['number'] == num:
            return doc['name']
    return None

def get_direct_by_num(num):
    for dir_num, doc_num_list in directories.items():
        if num in doc_num_list:
            return dir_num
    return None
        

def owner_document_by_number():
    a = input('Введите номер документа:\n') # a = '42'
    found_name = get_name_by_num(a)
    print('Результат:')
    if found_name is None:
        print('Документ не найден в базе')
    else:
        print('Владелец документа:', found_name)
        

def number_document_directories():
    b = input('Введите номер документа:\n') # a = '42'
    found_dir = get_direct_by_num(b)
    print ('Результат:')
    if not(found_dir is None):
        print('Документ хранится на полке:', found_dir)
    else:
        print('Документ не найден в базе' )  
        
def information_on_documents():
    print('Результат:')
    for doc_info in documents:
        shelf = get_direct_by_num(doc_info["number"])
        print(f'№: {doc_info["number"]}, тип: {doc_info["type"]}, '
              f'владелец: {doc_info["name"]}, полка хранения: {shelf}') 

#Функция на добавление
def add_shelves():
    print('Введите номер полки: ')
    x = input()
    print('Результат: ')
    if x in directories.keys():
        print('Такая полка уже существует.', end=' ')
    else:
        directories[x] = []
        print('Полка добавлена.', end=' ')
    print_shelves()            
    
#Функция на удаление 
def del_shelves():
        print('Введите номер полки: ')
        x = input()
        print('Результат: ')
        if x not in directories.keys():
            print('Такой полки не существует.', end=' ')
        elif len(directories[x]) > 0:
            print('На полке есть документы, удалите их перед удалением полки.', end=' ')
        else:
            del(directories[x])
            print("Полка удалена.", end=' ')
        print_shelves()
        
def print_shelves():
    print('Текущий перечень полок:', ', '.join(directories.keys()) + '.')
    
    
command = input('Введите команду:\n')
while command != 'q':
    if command == 'p':
        owner_document_by_number()
    elif command == 's':
        number_document_directories()
    elif command == 'l':
        information_on_documents()
    elif command == 'as':
        add_shelves()
    elif command == 'ds':
        del_shelves()
    else: 
        print("Неизвестная команда")
    command = input('Введите команду:\n')
```
